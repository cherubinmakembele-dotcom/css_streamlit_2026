# Problem 1
Write a program that reads integers from the user and stores them in a list. Your program should
continue reading values until the user enters 0. Then it should display all of the values entered
by the user (except for the 0) in ascending order, with one value appearing on each line. Use
either the sort method or the sorted function to sort the list.


```python
integers=[]
while True:
    user_input=int(input("enter a number: "))
    
    if user_input!=0:
        integers.append(user_input)
        integers.sort(reverse=False) # ⚠️⚠️⚠️⚠️
    else:
        break
    
print("\nyou have entered zero so the programs stops!")
print("\nthe list is: \n",integers)
        
    
```

# Problem 2
Write a program that reads integers from the user and stores them in a list. Use 0 as a sentinel
value to mark the end of the input. Once all of the values have been read your program should
display them (except for the 0) in reverse order, with one value appearing on each line.


```python
integers=[]
while True:
    user_input=int(input("enter a number: "))
    
    if user_input!=0:
        integers.append(user_input)
        integers.sort()
    else:
        break
    
for i in reversed(integers):
    print("\n",i)
        
    
```

# Problem 5
A proper divisor of a positive integer, n, is a positive integer less than n which divides evenly into
n. Write a program that computes all of the proper divisors of a positive integer. The integer will
be passed to the program as its only parameter. The program will return a list containing all of
the proper divisors as its only result.


```python
user_input=int(input("enter a number i'll tell you it's divisors: "))
diviseurs=[]
for i in range(1,user_input+1):
    if user_input%i==0:
        diviseurs.append(i)
print ("The proper divisors of",user_input,"are\n",diviseurs[:-1])
   
```

# Problem 6
An integer, n, is said to be perfect when the sum of all of the proper divisors of n is equal to n.
For example, 28 is a perfect number because its proper divisors are 1, 2, 4, 7 and 14, and 1 + 2+ 4 + 7 + 14 = 28. Write a program that determines whether or not a positive integer is perfect.
Your program will take one parameter. If that parameter is a perfect number then your program
will return True. Otherwise it will return False.


```python
def perfect(user_input):
    user_input=int(input("enter a number to test if it's perfect or not: "))
    diviseurs=[]
    for i in range(1,user_input+1):
        if user_input%i==0:
            diviseurs.append(i)
    if sum(diviseurs[:-1])==user_input:
        print(True,",",user_input, "is a perfect number")
    else:
        print(False,",",user_input, "is not a perfect number")

perfect(user_input)
```

# Problem 7
Write a program that reads numbers from the user until a blank line is entered. Your program
should display the average of all of the values entered by the user. Then the program should
display all of the below average values, followed by all of the average values (if any), followed
by all of the above average values. An appropriate label should be displayed before each list of
values

```python
numbers=[]
b_avg=[]
a_avg=[]
while True:
    user_input=input("enter negative and positive numbers: ")
    
    if user_input == "":
        break

    try:
        number = int(user_input)
        numbers.append(number)
        avg=sum(numbers)/len(numbers)
        
    except ValueError:
        print("\nInvalid input. Please enter a valid integer.")

for i in numbers:
    if i<avg:
        b_avg.append(i)
        b_avg.sort(reverse=False)
    if i>avg:
        a_avg.append(i)
        a_avg.sort(reverse=False)

print("\nthe avg of",numbers,"is \n",avg)
print("\nthe numbers below",avg,"of",numbers,"are \n",b_avg)
print("\nthe numbers above",avg,"of",numbers,"are \n",a_avg)


```
